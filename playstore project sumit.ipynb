## Data Cleaning of first Dataset **(Play store Data)**

## In **Size,Installs and Price** Columns

There are three columns values that we are able to use frequently henceforth are Installs, size, and Price. A cautious look of the dataset exhibits that a number of those columns need to be data cleaning .

The presence of special characters (, , $ , +) and letters (M , k) inside the Installs, Size, and Price in columns are object data type so that we convert these values into int/float data type . Their conversion to a numerical data type difficult.

## Cleaning **Size**

```python
# Replace 'Varies with device' value with 0.0......we will replace these values with mean value
# convert all values (like-'M','k') in one unit 'M'


df['Size']=df['Size'].replace('Varies with device',0.0)

df['Size']=df['Size'].apply(lambda x: str(x).replace('M','') if 'M' in str(x) else str(x))
df['Size']=df['Size'].apply(lambda x: str(x).replace('k','') if 'k' in str(x) else str(x))
df['Size']=df['Size'].apply(lambda x: float(x))
```

```python
# Replacing the 0.0 value with mean values
df['Size']=df['Size'].replace(0.0,df['Size'].mean())
```

## Cleaning **Installs**

```python
# Replace symbols and convert object data type to float data type 
df['Installs']=df['Installs'].apply(lambda x: str(x).replace('+','') if '+' in str(x) else str(x))
df['Installs']=df['Installs'].apply(lambda x: str(x).replace(',','') if ',' in str(x) else str(x))
df['Installs']=df['Installs'].apply(lambda x: float(x))
```

## Cleaning **Price** 

```python
# Check the unique value
df['Price'].unique()
```

```python
# convert into float value
df['Price']=df['Price'].apply(lambda x: str(x).replace('$','') if '$' in str(x) else str(x))
df['Price']=df['Price'].apply(lambda x: float(x))
```
